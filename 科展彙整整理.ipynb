```python
import pandas as pd
```

```python
df = pd.read_csv('科展資料彙整to2019 - 科展資料彙整.csv')  
print(df)  
```

```python
df.columns
```

```python
def make_clickable(link):
    # target _blank to open new window
    # extract clickable text to display for your link
    text = link
    return f'<a target="_blank" href="{link}">{text}</a>'

# link is the column with hyperlinks
df['檔案連結'] = df['檔案連結'].apply(make_clickable)
```

```python
result = df.sort_values(['屆次','科展類別'], ascending = [True, True])[['科展類別','屆次','科別','作品名稱']].to_html()
```

```python
with open('科展彙整.html', 'w', encoding='utf-8' ) as myFile:
    myFile.write('<html>')
    myFile.write('<head>')
    myFile.write('<title>')
    myFile.write('歷屆科展彙整至2019')    
    myFile.write('</title>')
    myFile.write('</head>')
        
    myFile.write('<body>')
    
    myFile.write( result )
    
    myFile.write('</body>')
   
    myFile.write('</html>')
```
